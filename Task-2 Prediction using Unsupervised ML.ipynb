### Question: From the given Iris Dataset, predict the optimum number of clusters and represent it visually

```python
# Importing all the required modules
import numpy as np

# For reading and storing data:
import pandas as pd

# For clustering, using Kmeans clustering model:
from sklearn.cluster import KMeans

# For showing the output
import matplotlib.pyplot as plt
```

```python
# Importing dataset
df = pd.read_csv("Task-2 Dataset Iris.csv")
df.head()
```

```python
# Taking x and y 
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Finding optimum number of clusters using elbow method
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Since there are 3 bends in the obtained graph, from elbow method, k = 3.

### Therefore the optimum number of clusters to be formed are "3".

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
# Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(nrows = 2, ncols = 2 , figsize=(15,10))

# Visualising the clusters - On the first two columns ie, between ID and sepal length
axis[0][0].scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
axis[0][0].scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
axis[0][0].scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'gray', label = 'Iris-virginica')

axis[0][0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'pink', label = 'Centroids')

axis[0][0].legend()

axis[0][0].set_title('K-Means Clustering Visualisation with Sepal Length')
axis[0][0].set_xlabel('ID')
axis[0][0].set_ylabel('Sepal Length (cm)')


# Visualising the clusters - On columns- 1 and 3 ie, between ID and sepal width
axis[0][1].scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 2], s = 100, c = 'purple', label = 'Iris-setosa')
axis[0][1].scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 2], s = 100, c = 'blue', label = 'Iris-versicolour')
axis[0][1].scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 2], s = 100, c = 'gray', label = 'Iris-virginica')

axis[0][1].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,2], s = 100, c = 'pink', label = 'Centroids')

axis[0][1].legend()

axis[0][1].set_title('K-Means Clustering Visualisation with Sepal Width')
axis[0][1].set_xlabel('ID')
axis[0][1].set_ylabel('Sepal Width (cm)')


# Visualising the clusters - On columns- 1 and 4 ie, between ID and petal length
axis[1][0].scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 3], s = 100, c = 'purple', label = 'Iris-setosa')
axis[1][0].scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 3], s = 100, c = 'blue', label = 'Iris-versicolour')
axis[1][0].scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 3], s = 100, c = 'gray', label = 'Iris-virginica')

axis[1][0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,3], s = 100, c = 'pink', label = 'Centroids')

axis[1][0].legend()

axis[1][0].set_title('K-Means Clustering Visualisation with Petal Length')
axis[1][0].set_xlabel('ID')
axis[1][0].set_ylabel('Petal Length (cm)')


# Visualising the clusters - On columns- 1 and 5 ie, between ID and petal width
axis[1][1].scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 4], s = 100, c = 'purple', label = 'Iris-setosa')
axis[1][1].scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 4], s = 100, c = 'blue', label = 'Iris-versicolour')
axis[1][1].scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 4], s = 100, c = 'gray', label = 'Iris-virginica')

axis[1][1].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,4], s = 100, c = 'pink', label = 'Centroids')

axis[1][1].legend()

axis[1][1].set_title('K-Means Clustering Visualisation with Petal Width (cm)')
axis[1][1].set_xlabel('ID')
axis[1][1].set_ylabel('Petal Width (cm)')

# Combine all the operations and display
figure.suptitle('Visualisation of Clusters with each feature', size = 20)
figure.subplots_adjust(top=0.88)
plt.tight_layout()
plt.show()
```
